```python
from abc import ABC, abstractmethod
class Payment:
  def __init__(self, customer_name, order_id, amount):
    self.customer_name = customer_name
    self.order_id = order_id
    self.amount = amount
    print("payment object created")
  @classmethod
  def pay(self):
    pass

  def payment_detail(self):
    print("------------------payment detail---------------------")
    print("customer_name                  : ", self.customer_name)
    print("order_id                       : ", self.order_id)
    print("amount                         : ", self.amount)

  def payment_status(self):
    print("payment_status                 : sucess")
    print("-----------------------------------------------------")

class Upi(Payment):
  def __init__(self, customer_name, order_id, amount, Upi_Id):
    super().__init__(customer_name, order_id, amount)
    self.Upi_Id = Upi_Id

  def pay(self):
    print("payment mode                   : upi")
    print("upi_id                         : ", self.Upi_Id)
    print(self.amount, "paid sucessfully using upi")

class Credit_card(Payment):
  def __init__(self, customer_name, order_id, amount, Credit_card_number):
    super().__init__(customer_name, order_id, amount)
    self.Credit_card_number = Credit_card_number

  def pay(self):
    print("payment mode                  : credit_card")
    print("credit_card_number            :  ",self.Credit_card_number)
    print("credit card payment sucessful : ", self.amount)


class Debit_card(Payment):
  def __init__(self, customer_name, order_id, amount, Debit_card_number):
    super().__init__(customer_name, order_id, amount)
    self.Debit_card_number = Debit_card_number

  def pay(self):
    print("payment mode                 : debit_card")
    print("debit_card_number            :  ",self.Debit_card_number)
    print("debit card payment sucessful : ", self.amount)

class Net_banking(Payment):
  def __init__(self, customer_name, order_id, amount, Net_banking_number):
    super().__init__(customer_name, order_id, amount)
    self.Net_banking_number = Net_banking_number

  def pay(self):
    print("payment mode                  : net_banking")
    print("net_banking_number            :  ",self.Net_banking_number)
    print("net banking payment sucessful : ", self.amount)

class Cash_on_delivery(Payment):
  def __init__(self, customer_name, order_id, amount):
    super().__init__(customer_name, order_id, amount)

  def pay(self):
    print("payment mode                  : cash_on_delivery")
    print(self.amount, "paid sucessfully using cash on delivery")
class Pay_later(Payment):
  def __init__(self, customer_name, order_id, amount):
    super().__init__(customer_name, order_id, amount)

  def pay(self):
    print("payment mode : pay_later")
    print(self.amount, "paid sucessfully using pay later")

name =         input("Enter your name               : ")
order_id1 =    input("Enter your order_id           : ")
amount1 =  int(input("Enter the pay amount          : "))



print("choose your payment option")
print("1. Upi")
print("2. Credit_card")
print("3. Debit_card")
print("4. Net_banking")
print("5. Cash_on_delivery")
print("6. Pay_later")
print("7. Exit")

choose = int(input("Enter payment option : "))

if choose == 1:
  print("you choose upi")
  A1 = Upi(name, order_id1, amount1, input("Enter your UPI ID : "))
  A1.payment_detail()
  A1.pay()
  A1.payment_status()
elif choose == 2:
  print("you choose credit_card")
  A2 = Credit_card(name, order_id1,amount1, input("Enter your credit card number : "))
  A2.payment_detail()
  A2.pay()
  A2.payment_status()
elif choose == 3:
  print("you choose debit_card")
  A3 = Debit_card(name, order_id1, amount1, input("Enter your debit card number : "))
  A3.payment_detail()
  A3.pay()
  A3.payment_status()
elif choose == 4:
  print("you choose net_banking")
  A4 = Net_banking(name, order_id1, amount1, input("Enter your net banking number : "))
  A4.payment_detail()
  A4.pay()
  A4.payment_status()
elif choose == 5:
  print("you choose cash_on_delivery")
  A5 = Cash_on_delivery(name, order_id1, amount1)
  A5.payment_detail()
  A5.pay()
  A5.payment_status()
elif choose == 6:
  print("you choose pay_later")
  A6 = Pay_later(name, order_id1, amount1)
  A6.payment_detail()
  A6.pay()
  A6.payment_status()
elif choose == 7:
  print("Exit")
else:
  print("Invalid choice")
```
